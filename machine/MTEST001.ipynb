```python
import pandas as pd
arq = pd.read_csv('C:/Users/thelsandro/Documents/GitHub/codigos_python/machine/wine_dataset.csv')
```

```python
arq.head()
```

```python
arq['style'] = arq['style'].replace('red',0)
```

```python
arq['style'] = arq['style'].replace('white',1)
```

```python
arq
```

```python
y = arq['style']
x = arq.drop('style', axis=1)
```

```python
arq.shape
```

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste,y_treino,y_teste = train_test_split(x,y,test_size=0.3)
```

```python
x_treino.shape
```

```python
from sklearn.ensemble import ExtraTreesClassifier

model = ExtraTreesClassifier()
model.fit(x_treino,y_treino)

result = model.score(x_teste,y_teste)
print('Acurácia', result)
```
